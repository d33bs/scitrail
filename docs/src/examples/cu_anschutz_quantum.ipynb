# CU Anschutz + Quantum Example

This notebook executes the built-in `scitrail` example using `institution: CU Anschutz` and `topic: Quantum`, then renders the generated report as rich notebook output.

```python
from pathlib import Path

from IPython.display import Markdown
from IPython.display import display as ipy_display

from scitrail.main import generate_example
```

```python
repo_root = Path.cwd()
if not (repo_root / "examples/cu_quantum.yaml").exists():
    repo_root = repo_root.parents[2]

config_path = repo_root / "examples/cu_quantum.yaml"
output_path = repo_root / "examples/cu_quantum_report.md"
report_path = generate_example(config_path=config_path, output_path=output_path)
report_text = report_path.read_text(encoding="utf-8")

report_path
```

```python
ipy_display(Markdown(report_text))
```
